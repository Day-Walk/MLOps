## tool1

```python
# parse_docstring=True : llm이 주석을 자동으로 파싱해서 이해
@tool()
def get_user_info(user_uuid):
  """
    사용자 정보 조회 도구

    주어진 user_uuid(사용자 고유 식별자)를 사용하여 데이터베이스에서 해당 사용자의 정보를 조회합니다.
    조회된 사용자 정보(나이, 성별, 선호 카테고리 등)를 딕셔너리 형태로 반환합니다.
    반환된 정보는 이후 맞춤형 추천 등 다음 도구의 입력값으로 활용할 수 있습니다.

    Args:
        user_uuid (str): 사용자 고유 식별자(예: 'b6e7...')

    Returns:
        dict: 사용자 정보
            - age (str): 사용자의 연령대 (예: '20대', '30대')
            - gender (str): 성별 (예: '남성', '여성')
            - like (str): 선호 카테고리 또는 키워드 (예: '맛집', '카페', '조용한')

    Example:
        >>> get_user_info("b6e7...")
        {'age': '20', 'gender': '남성', 'like': '카페'}
  """

  # 디비 조회

  user_info = {
      "age" : "20",
      "gender" : "남성",
      "like" : ['카페', '조용한', '깔끔한']
  }

  return user_info
```

## tool2

- tool2 : 엘라스틱 서치 검색

- 엘라스틱 서치 실행 함수를 호출

- in : llm이 추출한 키워드(카테고리, 지역)

- out : 장소uuid 리스트

```python
# 카테고리 리스트 사전 정의
categories = set()

for place in place_json:
    categories.add(place['category'])
    categories.add(place['subCategory'])
category_list = list(categories)
print(category_list)
```

```python
@tool
def elastic_search(query):

  """
    엘라스틱 검색 도구

    사용자가 원하는 지역(region)과 카테고리(categories)를 입력받아,
    해당 조건에 부합하는 장소들의 UUID 리스트를 반환합니다.
    만약 정확한 카테고리가 입력되지 않더라도, 유사한 카테고리를 추출하여 검색에 활용합니다.

    Args:
        query (dict): 검색 조건
            - region (str): 사용자가 명시한 지역명 (예: '강남', '홍대', '성수동', '이태원로', '잠실역')
            - categories (list of str): 최소 1개, 최대 3개의 장소 카테고리 (예: '전시관', '기념관', '전문매장/상가', '5일장', '특산물판매점', '백화점', '상설시장', '문화전수시설', '문화원', '서양식', '건축/조형물', '음식점&카페', '박물관', '컨벤션센터', '역사관광지', '복합 레포츠', '공예/공방', '이색음식점', '영화관', '산업관광지', '중식', '문화시설', '쇼핑', '수상 레포츠', '관광지', '육상 레포츠', '학교', '관광자원', '스키(보드) 렌탈샵', '대형서점', '휴양관광지', '외국문화원', '자연관광지', '레포츠', '한식', '일식', '도서관', '체험관광지', '카페/전통찻집', '면세점', '공연장', '미술관/화랑')

    Returns:
        list: 조건에 부합하는 장소의 UUID 리스트 (예: ['uuid1', 'uuid2', ...])

    Example:
        >>> elastic_search({'region': '종로', 'categories': ['음식점&카페', '문화시설']})
        ['a1b2c3', 'd4e5f6']
  """

  region = query['region']
  categories = query['categories']
  # 엘라스틱 서치 호출

  # 인코딩
  region_encoded = quote(region)
  category_encoded = '&'.join([f"categories={quote(cat)}" for cat in categories])

  # url
  url = f"http://15.164.50.188:9201/api/place/search/llm-tool?region={region_encoded}&{category_encoded}"

  # get 요청
  response = requests.get(url)
  # response.raise_for_status()

  # 응답 처리
  response_json = response.json()
  uuid_list = response_json.get('uuids', [])

  # 엘라스틱 서치 호출

  return uuid_list
```

## tool3

- tool3 : 동적 검색기 생성 및 실행

- 장소 uuid 리스트 안에 포함되는 장소 리스트를 먼저 프리필터링 하고

- 그 결과 내에서 임베딩 검색하기

- in : 장소 uuid 리스트

- out : 임베딩 결과 장소 리스트

```python
# 엘라스틱서치 검색 후 검색기 생성 및 실행
@tool
def search_with_filtering(data):
  """
    임베딩 검색 도구

    입력받은 UUID 리스트와 사용자의 쿼리문(query)을 기반으로,
    메타데이터 필터링 및 임베딩 유사도 검색을 수행하여 관련 장소 정보를 반환합니다.

    Args:
        data (dict): 검색 데이터
            - query (str): 사용자가 입력한 쿼리문 (예: '홍대에서 감성 카페 갔다가 전시 보는 코스 추천')
            - place_uuids (list of str): 검색 대상 장소의 UUID 리스트

    Returns:
        list: 임베딩 유사도 기반으로 필터링된 장소 정보 리스트

    Example:
        >>> search_with_filtering({'query': '감성적인 카페', 'place_uuids': ['uuid1', 'uuid2']})
        [{'name': 'OO레스토랑', 'address': '서울...', ...}, ...]
  """
  query = data['query']
  place_uuids = data['place_uuids']

  retriever = chroma_bge.as_retriever(
    search_type="similarity",
    search_kwargs={
        "k" : 100,
        "filter" : {
            "uuid" : {"$in": place_uuids}
        }
    }
  )

  place_list = retriever.invoke(query)

  return place_list
```
