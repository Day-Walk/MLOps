# image 얻어오기

```python
def get_image_list(contentid):
  image_list = []

  detailImage_url = 'http://apis.data.go.kr/B551011/KorService2/detailImage2'

  detailImage_params = {
      'serviceKey' : decoding_key,
      'MobileOS' : 'WEB',
      'MobileApp' : 'Sample',
      '_type' : 'json',
      'contentId' : contentid
  }
  detailImage_response = requests.get(detailImage_url, params=detailImage_params)
  detailImage = detailImage_response.json()
  image_items = detailImage['response']['body']['items']
  if image_items:
    image_item = image_items['item']
    image_list = [item['originimgurl'] for item in image_item]
  else:
    print("없음")

  return image_list
```
